# Covert string value attribute in numeric

```python
data.Attrition=data.Attrition.map({'Yes':1,'No':0})
```

```python
data.BusinessTravel=data.BusinessTravel.map({'Travel_Rarely':1, 'Travel_Frequently':2, 'Non-Travel':3})
```

```python
data.Department=data.Department.map({'Sales':1, 'Research & Development':2, 'Human Resources':3})
```

```python
data.EducationField=data.EducationField.map({'Life Sciences':1, 'Other':2, 'Medical':3, 'Marketing':4,
       'Technical Degree':5, 'Human Resources':6})
```

```python
data.Gender=data.Gender.map({'Female':0, 'Male':1})
```

```python
data.JobRole=data.JobRole.map({'Sales Executive':1, 'Research Scientist':2, 'Laboratory Technician':3,
       'Manufacturing Director':4, 'Healthcare Representative':5, 'Manager':6,
       'Sales Representative':7, 'Research Director':8, 'Human Resources':9})
```

```python
data.MaritalStatus=data.MaritalStatus.map({'Single':1, 'Married':2, 'Divorced':3})
```

```python
data.Over18=data.Over18.map({'Y':1})
```

```python
data.OverTime=data.OverTime.map({'Yes':1, 'No':0})
```

# Split independent variable and dependent variable

```python
x=data.iloc[:,2:]
y=data.iloc[:,1:2]
x0=data.iloc[:,:1]
```

```python
x=pd.concat([x0,x],axis=1,join='inner')
```

```python
x=x.values
y=y.values
```

```python
x
```

```python
y
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

# Split train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Logistic Regression Model

Train Model

```python
from sklearn.linear_model import LogisticRegression
log_r=LogisticRegression()
log_r.fit(x_train,y_train)
```

Test Model

```python
pre=log_r.predict(x_test)
```

```python
np.concatenate((y_test.reshape(len(y_test),1),pre.reshape(len(pre),1)),1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, pre)
print(cm)
accuracy_score(y_test, pre)
```

We goted the accuracy of our model is 87.8%

# K- nearest Neighbor Classification Model

Train Model

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=7,metric='minkowski',p=2)
knc.fit(x_train,y_train)
```

Test Model

```python
pre_knc=knc.predict(x_test)
```

```python
np.concatenate((y_test.reshape(len(y_test),1),pre_knc.reshape(len(pre_knc),1)),1)
```

```python
cm = confusion_matrix(y_test, pre_knc)
print(cm)
accuracy_score(y_test, pre_knc)
```

We goted the accuracy of our model is 84.35%
